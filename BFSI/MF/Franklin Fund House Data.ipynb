```python
df = pd.read_excel("Equity_factsheet_As_on_31_01_2019.xls", None);
```

```python
merged_data = {}
ignore_list = ['FIPP', 'FIEHF', 'FIESF']
for key in df.keys():
    #print(df[key].columns[0])
    if key not in ignore_list:
        print(f'Processing {key}')
        fund = df[key].columns[0]
        merged_data[key] = {}
        merged_data[key]['fund'] = fund
        headers = list(df[key].iloc[1])
        df[key].dropna(axis=0, how='all', thresh=None, subset=None, inplace=True)
        df[key] = df[key][2:]
        df[key].columns = headers
        df[key]['Sector'] = 'NA'
        sectype='NA'
        sectypeseries= []
        
        for row in df[key].iterrows():
            #print(row[1][0], row[1][4], type(row[1][1]), np.isnan(row[1][1]))
            if np.isnan(row[1][1]):
                sectype = row[1][0]
                #print(row[1][0])
            sectypeseries.append(sectype)
        df[key]['Sector'] = sectypeseries
        #sectypeseries
        df[key].dropna(inplace=True)
        #print(df[key])
        Total = df[key]['% of Assets'].sum()
        #print (Total)
        cash = 100 - Total
        df[key] = df[key].append({'Company Name':'Cash', '% of Assets':cash, 'Sector': 'Cash'}, ignore_index=True)
        merged_data[key]['data'] = df[key]
    else:
        print(f'Not Processing {key}')
```
